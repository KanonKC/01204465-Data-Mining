```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_excel('hw#2_1_data.xlsx')
```

```python
plt.hist(dataset['X2'], bins=4, label='X2',rwidth=0.99)
plt.title('Histrogram of X2')
```

```python
plt.hist(dataset['X3'], bins=3, label='X3', rwidth=0.99)
plt.title('Histrogram of X3')
```

```python
# Box plot for X3
plt.boxplot(dataset['X3'])
plt.title('Box plot of X3')
```

```python
# Scatterplot X3 and X4
plt.scatter(dataset['X3'], dataset['X4'], label='X3 and X4')
plt.title('Scatterplot of X3 and X4')
```
