```python
import pandas as pd
import plotly.express as px
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
#from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
#from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
```

# Import Data

```python
#datasets = pd.read_excel('Universal_bank_dataset.xls',engine='openpyxl',sheet_name=1)
datasets = pd.read_csv('Universal_bank_dataset.csv')
datasets.drop(['ID','ZIP Code'],axis=1,inplace=True)
```

```python
new_customer = pd.read_csv('Universal_bank_new.csv')
new_customer.drop(['Personal Loan'],axis=1,inplace=True)
new_customer
```

```python
y_df = datasets[['Personal Loan']]
datasets.drop(['Personal Loan'],axis=1,inplace=True)
X_df = datasets
with pd.option_context('display.max_rows', 8): display(X_df)
```

```python
X = X_df.values
y = y_df.values
```

# Preprocess

```python
scaler = StandardScaler()
X_norm = scaler.fit_transform(X)
new = new_customer.values
new = scaler.transform(new)
```

# K-NN k=5

```python
neigh = KNeighborsClassifier(n_neighbors=5,metric='euclidean')
neigh.fit(X_norm, y)
```

```python
y_pred = neigh.predict(new)
print(y_pred)
```

# Find Best k K-NN

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,train_size=0.6,random_state=1992)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
scores = []
for k in range(1,21):
    neigh = KNeighborsClassifier(n_neighbors=k,metric='euclidean')
    neigh.fit(X_train, y_train.ravel())
    scores.append([k,neigh.score(X_test,y_test)])
```

```python
scores
```

```python
scores.sort(key=lambda x: x[1], reverse=True)
print(f"Best k={scores[0][0]} with accuracy score:{scores[0][1]}")
```

# Using Best k & Show Confusion Matrix

```python
k = scores[0][0]
neigh = KNeighborsClassifier(n_neighbors=k,metric='euclidean')
neigh.fit(X_train, y_train.ravel())
```

```python
y_pred = neigh.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
