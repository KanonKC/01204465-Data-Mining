```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from imblearn.over_sampling import KMeansSMOTE
from sklearn.datasets import make_blobs
```

```python
blobs = [100, 800, 100]
X, y  = make_blobs(blobs, centers=[(-10, 0), (0,0), (10, 0)])

sns.scatterplot(x=X[:,0], y=X[:,1],hue=y)
```

```python
# Make this a binary classification problem
sm = KMeansSMOTE(k_neighbors=3)
X_res, y_res = sm.fit_resample(X, y)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, ax=ax1).set_title('Original Data')
sns.scatterplot(x=X_res[:, 0], y=X_res[:, 1], hue=y_res, ax=ax2).set_title('KMeansSMOTE Oversampling')
```

```python
# Plot 10 sns scatterplots with different k_neighbors
fig, ax = plt.subplots(5,2, figsize=(10, 25))
for i in range(10):
    sm = KMeansSMOTE(k_neighbors=i+1)
    X_res, y_res = sm.fit_resample(X, y)
    sns.scatterplot(x=X_res[:, 0], y=X_res[:, 1], hue=y_res, ax=ax[ i%5,i//5]).set_title(f'KMean Smote with k={i+1}')
```
